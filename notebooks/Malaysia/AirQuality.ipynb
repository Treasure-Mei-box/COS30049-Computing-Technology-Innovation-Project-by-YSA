```python
import pandas as pd

file_path = "/content/drive/MyDrive/AQIs/langkawi,-kedah-air-quality.csv"
df = pd.read_csv(file_path)

# Clean column names
df.columns = df.columns.str.strip().str.replace(" ", "_")

# Keep only Date & AQI columns (if names differ, print df.columns to check)
df = df[['date', 'aqi']]

# Convert Date to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Filter only 2014–2024
df = df[df['date'].dt.year.between(2014, 2024)]

# Sort ascending by date (oldest → newest)
df = df.sort_values('date')

# Save cleaned file
cleaned_path = "/content/drive/MyDrive/AQIs/langkawi_cleaned_2014_2024.csv"
df.to_csv(cleaned_path, index=False)

print("Cleaned data saved to:", cleaned_path)
print(df.head())
```
